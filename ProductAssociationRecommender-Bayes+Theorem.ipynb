```python
# empty dataframe for movie-movie afiniti score
movie_afiniti = pd.DataFrame(columns=[
    'base_movieId',
    'base_movieTitle',
    'associated_movieId',
    'associated_movieTitle',
    'afiniti_score'])
```

```python
#Write a function to calcuate the affinity score for given movie id
def get_Score(movieid):
    
    Affiniti_score_movies=[]
    Affiniti_dict={}
    
    
    
    #Get the user ids for base movie
    Base_movie=ratings[ratings['movieId']==movieid]
    distinctuserid_m1=np.unique(Base_movie['userId'])
    
    #print("m1 viewers",distinctuserid_m1)
        
        
    for m2 in distinct_movies:
              
        if m2==movieid:
            continue
                
            #users for movie # 2
        x=ratings[ratings['movieId']==m2]
        distinctuserid_m2=np.unique(x['userId'])
        
        #print("m1 viewers",distinctuserid_m2)
        
        
        #Get the intersecting users
        intersection_users=np.intersect1d(distinctuserid_m1,distinctuserid_m2)
        
        #Get the count of common viewers
        intersection_users_count=len(intersection_users)
        #print the intersection count
        #print("Common users",intersection_users_count)
        
        #Calculate the affinity score
        affinity_Score=float(intersection_users_count)/len(distinctuserid_m1)
        
        Affiniti_score_movies.append(affinity_Score)
        
        Affiniti_dict[m2]=affinity_Score
        
        #print("Affinity _Score between " + " " + str(movieid) + "  and" + " " + str(m2) + " " + str(affinity_Score))   
        
        
        
        #print(Affiniti_score_movies)
        #Append the movie affiniti details    
        #movie_afiniti = movie_afiniti.append({
        #      "base_movieId": ref_movie,
        #      "base_movieTitle": ratings[ratings['movieId']==movieid].iloc[0,3],
        #      "associated_movieId": m1,
        #      "associated_movieTitle": ratings[ratings['movieId']==m2].iloc[0,3],
        #      "afiniti_score": affinity_Score}, ignore_index=True)    
        
    Affiniti_score_movies=sorted(Affiniti_score_movies,reverse=True)
    
    print(Affiniti_score_movies)
    
    return Affiniti_dict  
    #Use threshold of 0.3 to filter only similar movies
       
        
             
   
    
    
    
```

```python
#Run the affiniti score calculation for movie id 
movieid=2
Affiniti_dict=get_Score(movieid)
```
